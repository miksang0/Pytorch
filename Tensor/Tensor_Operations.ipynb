```python
import torch
```

```python
my_tensor = torch.arange(15)
my_tensor
```

```python
#Reshape and View: It helps us reshape the data.
my_tensor = my_tensor.reshape(3, 5)
my_tensor
```

```python
#Reshaping when we don't know the items.
# -1 helps us determine it during this case.
my_tensor2 = torch.arange(15)
my_tensor2
```

```python
my_tensor2 = my_tensor.reshape(3, -1)
my_tensor2
```

```python
#View
my_tensor3 = torch.arange(15)
my_tensor3
```

```python
my_tensor3 = my_tensor3.view(3, 5)
my_tensor3
```

```python
#Reshape and View will update automatically.
my_tensor4 = torch.arange(10)
my_tensor4
```

```python
my_tensor5 = my_tensor4.reshape(2, 5)
my_tensor5
```

```python
my_tensor4[2] = 45
my_tensor4
```

```python
my_tensor5
```

```python
#Slices
my_tensor6 = torch.arange(10)
my_tensor6
```

```python
my_tensor6[5]
```

```python
#Grabbing a slice
my_tensor7 = torch.arange(15)
my_tensor7
```

```python
my_tensor7 = my_tensor7.reshape(5, 3)
my_tensor7
```

```python
my_tensor7[:, 1]
```

```python
#Returning the column(s)
my_tensor7[:, 1:2]
```

**Tensor Math Operations**

```python
tensor_a = torch.tensor([1, 2, 3, 4])
tensor_b = torch.tensor([5, 6, 7, 8])
```

```python
#Addition
tensor_a + tensor_b
```

```python
#Addition Function
torch.add(tensor_a, tensor_b)
```

```python
#Subtraction
tensor_b - tensor_a
```

```python
#Subtraction Function
torch.sub(tensor_b, tensor_a)
```

```python
#Multiplication
tensor_a * tensor_b
```

```python
#Multiplication Function
torch.mul(tensor_a, tensor_b)
```

```python
#Division
tensor_b / tensor_a
```

```python
#Division Longhand
torch.div(tensor_b, tensor_a)
```

```python
#Remainder
tensor_b % tensor_a
```

```python
#Remainder Function
torch.remainder(tensor_b, tensor_a)
```

```python
#Power
torch.pow(tensor_a, tensor_b)
```

```python
#Reassignment
#tensor_a = tensor_a + tensor_b #Option 1#

tensor_a.add_(tensor_b)   #Option 2
tensor_a
```
